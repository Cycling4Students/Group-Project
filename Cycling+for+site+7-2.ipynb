```python
import pandas as pd
d7ch1= pd.read_csv('07 - 17.01.29-02.04 ch1.csv', delimiter=';', quotechar='|')
d7ch1.head()
```

I set titles for the columns before cleaning data.

```python
d7ch1.columns = (['Hour','PC', 'KpH', 'Ped', 'Total'])
d7ch1
```

The column 'Ped' is irrelevant for my anaysis so I will drop the 'Ped' and 'Total' columns:

```python
da7ch1 = d7ch1.drop('Ped', axis =1)
db7ch1 = da7ch1.drop('Total', axis =1)
db7ch1
```

```python
#further cleaning of irrelevant rows
dc7ch1 = db7ch1[d7ch1.Hour != 'SITE NO:']
dd7ch1 = dc7ch1[d7ch1.Hour != 'Date:']
de7ch1 = dd7ch1[d7ch1.Hour != 'Direction:']
df7ch1 = de7ch1[d7ch1.Hour != 'Total']
dg7ch1 = df7ch1[d7ch1.Hour != 'Hour']
dg7ch1
```

```python
dg7ch1.dropna(how='all', inplace=True) #I removed the blank rows.
dg7ch1
```

```python
dg7ch1['Hours'] = dg7ch1.Hour.str.split('-').str.get(0)
dg7ch1['Hourend'] = dg7ch1.Hour.str.split('-').str.get(1)
dg7ch1.drop(dg7ch1.columns[0], axis=1, inplace=True)


pd.options.mode.chained_assignment = None
```

```python
dg7ch1[['Hours','Hourend','PC', 'KpH']] = dg7ch1[['Hours','Hourend','PC', 'KpH']].apply(pd.to_numeric)

d7ch1day1 = dg7ch1.iloc[0:24] #I create a dataframe for each day.
d7ch1day2 = dg7ch1.iloc[24:48]
d7ch1day3 = dg7ch1.iloc[48:72]
d7ch1day4 = dg7ch1.iloc[72:96]
d7ch1day5 = dg7ch1.iloc[96:120]
d7ch1day6 = dg7ch1.iloc[120:144]
d7ch1day7 = dg7ch1.iloc[144:168]

d7ch1day1
```
