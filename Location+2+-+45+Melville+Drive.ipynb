```python
import pandas as pd
data = pd.read_csv("45-17.01.29-02.04.csv")
print(data)
```

```python
data = data.rename(columns={'Unnamed: 1': 'Hour', 'Unnamed: 2': 'Bike Count', 'Unnamed: 3': 'Bike Speed','DAILY FLOW BY HOUR': 'Ped'})
print('45 Melville Drive - DAILY FLOW BY HOUR')
data.drop(data.index[[0,1,2,3,4,5]], inplace=True)
data.drop(data.columns[[0,5]], axis=1, inplace=True)
data.index = data.index - 6
newcols = {
    'P/C': 'Bike Counts', 
    'P/C Kph': 'Bike Speed', 
    'Ped': 'Pedestrian Counts'
}
data.rename(columns=newcols, inplace=True)
data
```

```python
data['Hourstarts'] = data.Hour.str.split('-').str.get(0)
data['Hourends'] = data.Hour.str.split('-').str.get(1)
```

```python
data.drop(data.columns[0], axis=1, inplace=True)
```

```python
data = data.dropna(how='all')
data
```

```python
data = data[data.Hourstarts != "Direction:"]
data
```

```python
data = data[data.Hourstarts != "Total"]
data
```

```python
data = data[data.Hourstarts != "Hour"]
data
```

```python
data = data[data.Hourstarts != "Date:"]
data
```

```python
pd.options.mode.chained_assignment = None
```

```python
data[['Bike Count','Hourstarts','Hourends','Bike Speed','Pedestrian Counts']] = data[['Bike Count','Hourstarts','Hourends','Bike Speed','Pedestrian Counts']].apply(pd.to_numeric)
#
```

```python
data.reset_index(inplace=True)
```
