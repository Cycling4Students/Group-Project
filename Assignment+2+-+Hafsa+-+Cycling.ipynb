__ABOUT MY DATA FILES__
* I had 33 files about Cycling in Edinburgh in __xsl__ format with four channels for each. Each file consisted of __hourly data__ for a counter on a different location in Edinburgh for __a week__ (from 29 Jan to 4 Feb 2017).
* The first two channels included number of bikes and their speeds on __different (two) directions__ at the same location.

__AIM__

I wanted to explore the relationship between site and cycling by understanding: 
* if there is a correlation between the level of speed and the number of bikes passing through certain locations (maybe people choose sites where they can cycle with higher level of speed).
* if there are differences in cycling rates for weekdays and weekends, daytime and nighttime (which could inform further site analysis).

For this assignment, I avoided other information about the sites from other sources but only considered the data I have in the files to make hypotheses about those relationships.

__STEPS__
* I chose __6 sites__ that are close to Meadows and the University of Edinburgh Main Campus to avoid a wider range of variants (i.e. site properties, people profiles).
* I have converted the two relevant channels of the files into separate csv files.
* After data cleaning and wrangling, I have made several data analysis and visualizations regarding the directions, daytime-nighttime and weekday-weekend rates to come up with several hypotheses mostly based on site 22.
* I made calculations for all 6 sites for further data analysis.
